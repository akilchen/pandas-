# **1.6 Column的新增及刪除**

```python
movies = pd.read_csv('/content/drive/MyDrive/Pandas/data/movie.csv')
```

02 assign():新增column

```python
col_map = {
      'aspect_ratio':'aspect',
      'movie_facebook_likes':'fb_likes'
      }
movies = movies.rename(columns=col_map).assign(has_seen=0)
movies.head(3)
```

03 將所有的演員及導演的fb按讚數作加總，放在total_likes(新的)欄位。

```python
total = (movies['actor_1_facebook_likes']+
        movies['actor_2_facebook_likes']+
        movies['actor_3_facebook_likes']+
        movies['director_facebook_likes'])
total.head(5)
```

```python
cols=['actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes','director_facebook_likes']
sum_col=movies[cols].sum(axis='columns') #用SUM()的方式讓missing value=0
sum_col.head(5)
```

```python
movies.assign(total_likes=sum_col).head(5)
```

05 計算cast_total_facebook_likes占total_likes的比例<br>
(1)先驗證cast_total_facebook_likes是否大於total_likes。

```python
#1.在movies DataFrame中新增一個'total_likes'欄位，表示電影的總按讚數。
#2.在movies DataFrame中新增一個'is_cast_likes_more'欄位，表示演員的Facebook總按讚數是否大於或等於電影的總按讚數。
#3.將新增了這兩個欄位的DataFrame儲存在 df2 變數中。

def cast_like_gt_actor_director(df):
  return df['cast_total_facebook_likes'] >= df['total_likes'] #比較DataFrame中每一列的'cast_total_facebook_likes'值是否大於'total_likes'值
df2 = (movies.assign(total_likes=total,
           is_cast_likes_more=cast_like_gt_actor_director))
```

(2)用all()來確認這個column裡所有的直都是True

```python
df2['is_cast_likes_more'].all()
```

(3)透過drop()刪除total_likes欄位，再重新建立不含導演的按讚數的Series。

```python
df2 = df2.drop(columns='total_likes')
#使用for變數in list讀取list，篩選出col_name同時有actor和_likes的column
actor_sum = (movies[[c for c in movies.columns if 'actor'in c and '_likes' in c]]
#沿著欄位方向（即水平方向）進行加總
       .sum(axis='columns'))

actor_sum.head(5)
```

(4)使用>=或ge()再檢查，看看cast_total_facebook_likes是否都大於actor_sum的值

```python
movies['cast_total_facebook_likes'].ge(actor_sum).all()
```

(5)計算cast_total_facebook_likes在actor_sum的%

```python
pct_like = (actor_sum.div(movies['cast_total_facebook_likes']))
pct_like.head(5)
```

```python
pct_like.describe()
```
