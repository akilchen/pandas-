# **0.2 DataFrame的屬性**

01 提取DataFrame物件的columns(欄位名稱)、index(索引標籤)及data(資料)

```python
movies = pd.read_csv('/content/drive/MyDrive/Pandas/data/movie.csv')

columns = movies.columns #取得columns
index = movies.index   #取得index
data = movies.values   #取得data
```

04 index和columns皆屬Index的子類別，這代表後續可以使用類似的操作處理它們

```python
issubclass(columns.__class__,pd.Index)  #使用issubclass()進行檢查
```

```python
issubclass(index.__class__,pd.Index)  #使用issubclass()進行檢查
```
