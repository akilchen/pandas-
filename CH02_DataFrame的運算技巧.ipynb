# **2.3 排序col_name**

01 讀取movies.csv,整理部分col_name,讓它更簡單

```python
movies = pd.read_csv('/content/drive/MyDrive/Pandas/data/movie.csv')
def shorten(col_name):
  return (col_name.replace('facebook_likes','fb').replace('for_reviews',''))

movies = movies.rename(columns=shorten)
movies.columns
```

03 更據欄位類型進行分類，進而產生多個col_name的分組list

```python
cate_core = ['movie_title', 'title_year','content_rating','genres']
cate_people = ['director_name','actor_1_name','actor_2_name','actor_3_name']
cate_other = ['color','country', 'language', 'plot_keywords', 'movie_imdb_link']

cont_fb = ['director_fb','actor_1_fb','actor_2_fb','actor_3_fb','cast_total_fb','movie_fb']
cont_finance =['budget','gross']
cont_num_reviews =['num_voted_users','num_user_','num_critic_']
cont_other =[ 'imdb_score','duration', 'aspect_ratio','facenumber_in_poster']
```

04 更據自己希望的方式將各組別串接起來，使用set()

```python
new_col_order = cate_core+cate_people+cate_other+cont_fb+cont_finance+cont_num_reviews+cont_other
set(movies.columns) == set(new_col_order)
```

05 將從新排好的col_name列出新的DataFrame

```python
movies[new_col_order].head()
```

# **2.4 DataFrame的統計方法**

01 讀取movie.csv,用shape、size、ndim取得其基本屬性，之後執行len()取得列數

```python
movies = pd.read_csv('/content/drive/MyDrive/Pandas/data/movie.csv')
print("shape=",movies.shape) #列數、行數(傳回tuple)
print('size=',movies.size)  #資料總數(4916*28)
print('ndim=',movies.ndim)  #維度數(軸數axis)
print('rows=',len(movies))  #列數
print('columns=',len(movies.columns))  #行數
```

02 用count()顯示每一個欄位的no missing values,並output每一個欄位為index的Series:

```python
movies.count()
```

03 min(),max(),mean(),median()和std()等統計方法會傳回一個Series。

```python
movies.select_dtypes(include=np.number).min()
```

```python
max = movies.select_dtypes(include=np.number).max()
formatted_max = max.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
formatted_max
```

```python
mean = movies.select_dtypes(include=np.number).mean()
formatted_mean = mean.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
formatted_mean
```

```python
median = movies.select_dtypes(include=np.number).median()
formatted_median = median.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
formatted_median
```

```python
std = movies.select_dtypes(include=np.number).std()
formatted_std = std.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
#lambda x: f"{x:.2f}" 是一個匿名函式，它會將每個元素 x 格式化為小數點後兩位的浮點數字串。
formatted_std
```

04 用describe()進行描述性統計資料及四分位數。T為轉置攔位。<br>
<br>
Gemini葛格說:
apply() 適用於對整行或整列進行操作，例如計算總和、平均值等。<br>
applymap() 適用於對 DataFrame 中的每一個元素進行操作，例如格式化、轉換等。<br>
在這個情況下，我們需要對 DataFrame 中的每一個數值進行格式化，因此 applymap() 是更適合的選擇。

```python
described_movies = movies.describe().T

# 使用applymap()和字串格式化，將每個元素都格式化為小數點後兩位
formatted_described_movies = described_movies.applymap(lambda x: f"{x:.2f}")
formatted_described_movies
```

05 describe()預設顯示四分位數，而我們也能用pencentiles來指定其他分位數。

```python
described_movies = movies.describe(percentiles=[.99]).T

# 使用applymap()和字串格式化，將每個元素都格式化為小數點後兩位
formatted_described_movies = described_movies.applymap(lambda x: f"{x:.2f}")
formatted_described_movies
```
