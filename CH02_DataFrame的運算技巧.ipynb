# **2.4 DataFrame的統計方法**

01 讀取movie.csv,用shape、size、ndim取得其基本屬性，之後執行len()取得列數

```python
movies = pd.read_csv('/content/drive/MyDrive/Pandas/data/movie.csv')
print("shape=",movies.shape) #列數、行數(傳回tuple)
print('size=',movies.size)  #資料總數(4916*28)
print('ndim=',movies.ndim)  #維度數(軸數axis)
print('rows=',len(movies))  #列數
print('columns=',len(movies.columns))  #行數
```

02 用count()顯示每一個欄位的no missing values,並output每一個欄位為index的Series:

```python
movies.count()
```

03 min(),max(),mean(),median()和std()等統計方法會傳回一個Series。

```python
movies.select_dtypes(include=np.number).min()
```

```python
max = movies.select_dtypes(include=np.number).max()
formatted_max = max.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
formatted_max
```

```python
mean = movies.select_dtypes(include=np.number).mean()
formatted_mean = mean.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
formatted_mean
```

```python
median = movies.select_dtypes(include=np.number).median()
formatted_median = median.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
formatted_median
```

```python
std = movies.select_dtypes(include=np.number).std()
formatted_std = std.apply(lambda x: f"{x:.2f}")  #使用字串格式化來控制輸出的格式，:.2f 表示將數值格式化為小數點後兩位的浮點數。
#lambda x: f"{x:.2f}" 是一個匿名函式，它會將每個元素 x 格式化為小數點後兩位的浮點數字串。
formatted_std
```

04 用describe()進行描述性統計資料及四分位數。T為轉置攔位。<br>
<br>
Gemini葛格說:
apply() 適用於對整行或整列進行操作，例如計算總和、平均值等。<br>
applymap() 適用於對 DataFrame 中的每一個元素進行操作，例如格式化、轉換等。<br>
在這個情況下，我們需要對 DataFrame 中的每一個數值進行格式化，因此 applymap() 是更適合的選擇。

```python
described_movies = movies.describe().T

# 使用applymap()和字串格式化，將每個元素都格式化為小數點後兩位
formatted_described_movies = described_movies.applymap(lambda x: f"{x:.2f}")
formatted_described_movies
```

05 describe()預設顯示四分位數，而我們也能用pencentiles來指定其他分位數。

```python
described_movies = movies.describe(percentiles=[.99]).T

# 使用applymap()和字串格式化，將每個元素都格式化為小數點後兩位
formatted_described_movies = described_movies.applymap(lambda x: f"{x:.2f}")
formatted_described_movies
```

```python
colleges = pd.read_csv('/content/drive/MyDrive/Pandas/data/college.csv',index_col='INSTNM')
college_ugds = colleges.filter(like='UGDS_')
college_ugds.head()
```

01 對college_ugds找出Northwest-Shoals Community College這一列的資料UGDS_資料。<br>
Pandas採用bankers_rounding的方法做數值進位()

```python
name = 'Northwest-Shoals Community College'
college_ugds.loc[name] #選擇這個欄位的其中一列的資料
```

同場加映:如果要讓他成為到小數點第二位，四捨五入的話，要轉換成Decimal格式並使用ROUND_HALF_UP模式進行。

```python
from decimal import Decimal, ROUND_HALF_UP

#college_ugds.loc[name]是一個Series
series = college_ugds.loc[name]

#將Series中的數值四捨五入到小數第二位
rounded_series = series.apply(lambda x: float(Decimal(str(x)).quantize(Decimal('0.00'), rounding=ROUND_HALF_UP)))
#  decimal.Decimal(str(x))： 將數值轉換為Decimal物件，以確保精確的十進位運算。
#  .quantize(Decimal('0.00'), rounding=ROUND_HALF_UP)：將Decimal物件量化到小數第二位，並使用ROUND_HALF_UP模式進行四捨五入。
#  float(...)：將結果轉換回浮點數。
#  series.apply(...)：將函數應用於Series中的每個元素。
#  df.applymap(...)：將函數應用於DataFrame中的每個元素。

rounded_series
```

# **2.8 DataFrame轉置(軸)計算**

許多DataFrame的方法都有一個**axis參數**，用來控制計算方向。axis參數預設為0(直向計算)，如果要橫向計算的話，axis=1。計算出結果以Series傳回。

01 讀取college.csv資料集並使用filter()選取特定欄位

```python
college = pd.read_csv('/content/drive/MyDrive/Pandas/data/college.csv',index_col='INSTNM')
college_ugds = college.filter(like='UGDS_')
college_ugds.head()
```

02 用count()傳回每個column no missing value的數量。

```python
college_ugds.count()
```

03 將axis參數設為1(columns),會改變計算方向，output每列no missing value的數量。

```python
college_ugds.count(axis=1).head()
```

04 對每一列中的所有直求和。由於每一個欄位為不同種族的百分比(單選)，因此每列總和應該是1，用sum()驗證這點~

```python
college_ugds.sum(axis=1).head() #橫向相加
```

05 使用median()了解每一個欄位的分布狀況

```python
college_ugds.median()
```

了解更多:cumsum()在axis=1中會沿著方向累加。

```python
college_ugds = college_ugds.cumsum(axis=1)
college_ugds.head()
```
