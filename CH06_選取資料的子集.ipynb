# **6.1 選取一筆或多筆Series資料**

01 讀取college.csv,用[]選取名為CITY的col_name,傳回Series物件存成city變數

```python
college = pd.read_csv('/content/drive/MyDrive/Pandas/data/college.csv',index_col='INSTNM')
city = college['CITY']
city
```

02 直接從傳回的Series中，用[]按照index label來選取資料

```python
city['Alabama A & M University']
```

03 使用loc按index label來做同樣的事情

```python
city.loc['Alabama A & M University']
```

04 使用iloc按照位置屬性作相同的事情

```python
city.iloc[0]
```

05 將多個包含index label的list放進[]中，就可以同時選取多筆資料。

```python
city[['Alabama A & M University','Florida State University']]
```

06 使用iloc做相同的事

```python
city.iloc[[0,1]]
```

08 使用[]搭配slicing來選取多個值。

```python
city['Alabama A & M University':'Alabama State University']
```

09 使用[]搭配位置slicing選取多個值

```python
city[0:5]
```

10 使用loc[]搭配label slicing做相同的事情

```python
city.loc['Alabama A & M University':'Alabama State University']
```

11 使用iloc[ ] 搭配位置slicing來做相同的事

```python
city.iloc[0:5]
```

12 用isin()判斷城市座落在Birmingham或Montgomery的大學，傳回一個boolean series

```python
alabama_mask = city.isin(['Birmingham','Montgomery'])
alabama_mask
```

```python
city[alabama_mask]
```

# **6.2 選取DataFrame的row**

要選取DataFrame的列，最直接的方法是使用iloc與loc

01讀取college.csv,將INSTNM設為index

```python
college = pd.read_csv('/content/drive/MyDrive/Pandas/data/college.csv',index_col='INSTNM')
college.head()
```

02 用iloc[index位置]取得特定資料

```python
college.iloc[0]
```

03 用loc[index label]取得資料

```python
college.loc['Amridge University']
```

04 如果要選取多列資料，以list包起不同列的位置，傳給iloc

```python
college.iloc[[60,178,935]] #會傳回一個DataFrame
```

05 將不同大學名稱(index label)存進list並傳遞給loc,有相同效果

```python
labels = ['Amridge University','Alabama A & M University','Alabama State University']
college.loc[labels]
```

06 使用iloc[slicing]來選取連續資料列

```python
college.iloc[99:102]
```

07 使用loc[slicing]來選取連續資料列<br>
(注意!搭配loc時，Pandas是使用closed interval)

```python
start = 'Alabama A & M University'
stop = 'Alabama State University' #傳回結果會包含結束的值
college.loc[start:stop]
```
