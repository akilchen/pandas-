# **6.4 混用loc與iloc來選取資料**

01 讀取college.csv，指定INSTNM為index

```python
college = pd.read_csv('/content/drive/MyDrive/Pandas/data/college.csv',index_col='INSTNM')
```

02 Chainning column和get_loc()求出所需欄位的整數位置。底下分別求出起始欄位和結束欄位的整數位置，以利用於slicing運作

```python
col_start=college.columns.get_loc('UGDS_WHITE')   #求出UGDS_WHITE欄位數字
col_end =college.columns.get_loc('UGDS_UNKN')+1  #求出UGDS_UNKN欄位數字，並將傳回結果加上1(closed interval)
col_start,col_end
```

03 使用col_start和col_end,搭配iloc來選取資料

```python
college.iloc[:5,col_start:col_end]
```

了解更多:使用index求出相應label，改用loc來選取

```python
row_start = college.index[10]
row_end = college.index[15]
row_start,row_end
```

```python
college.loc[row_start:row_end,'UGDS_WHITE':'UGDS_UNKN']
```

# **6.5 按標籤字母的順序進行slicing**

01 讀取college,csv,指定INSTNM為index

```python
college = pd.read_csv('/content/drive/MyDrive/Pandas/data/college.csv',index_col='INSTNM')
```

02 選取index label在Sp和Su之間的大學

```python
college.loc['Sp':'Su']
#在index未排序狀況下，loc只能根據傳遞給它的確切標籤來選取資料。當DF index找不到這些label時，就會error。
```

03 用sort_index()排序index label

```python
college= college.sort_index()
```

```python
college.loc['Sp':'Su'] #不含Su
```

了解更多:index label按相反方向排序，也能使用slicing。可以使用is_monotonic_increasing或is_monotonic_decreasing來判斷index方向<br>
以下將index label做逆向排序

```python
college = college.sort_index(ascending=False) #逆向排序
college.index.is_monotonic_decreasing #判斷index方向是不是逆向
```

```python
college.loc['E':'B'] #不包含Exx,因為Exx在E前面
```

# **Q:練習題：**

(1)選出「城市是新北」的所有客戶<br>
(2)選出「貸款金額超過 600000」的客戶名字和金額<br>
(3)試試用 .query() 寫出上面第 2 題的方式<br>

```python
import pandas as pd

data = {
    '客戶': ['小志', '小明', '小美', '阿凱', '阿德'],
    '城市': ['台北', '台北', '新北', '新北', '台中'],
    '貸款金額': [500000, 800000, 600000, 600000, 700000],
    '放款日': ['2023-01-10', '2023-02-15', '2023-01-20', '2023-03-01', '2023-01-05']
}
df = pd.DataFrame(data)
df
```

```python
df.loc[df['城市'] == '新北']
```

```python
df2 = df.sort_values('貸款金額')
df2
```

```python
over6 = df[df['貸款金額'].gt(600000)][['客戶','貸款金額']]
over6
```

```python
df['貸款金額']
```

```python
##3試試用 .query() 寫出上面第 2 題的方式
df2 = df.query('貸款金額 > 60000')
df2
```

```python
df3=df.query('貸款金額 > 600000')[['客戶', '貸款金額']]
df3
```
